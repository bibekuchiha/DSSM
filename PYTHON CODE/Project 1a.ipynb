## 1. Import an XML file using python

```python
# importing element tree
import xml.etree.ElementTree as ET 

# Pass the path of the xml document 


tree = ET.parse("movies.xml") 
```

```python
# get the parent tag 
root = tree.getroot() 
```

```python
# print the root (parent) tag along with its memory location 
print(root)  
```

```python
# print the attributes of the first tag  
print(root[0].attrib) 
```

```python
#print all the attributes of the tags
for i in root:
  print(i.attrib)
```

```python
#printing the details of each tag
for i in root:
  print(i.attrib)
  for j in i:
    print("\t",j.tag,":",j.text)
```

## 3. Import the breast cancer dataset from sklearn library and attach the target variable data to the features data and store it in a JSON file

```python
import pandas as pd

#importing the Sklearn Dataset breast_cancer
from sklearn.datasets import load_breast_cancer

#importing data
data = load_breast_cancer()

#getting the list of features and target in the dataset
print("The Features in the dataset are : \n",data.feature_names)
print("The Target is : \n",data.target_names)
```

```python
#creating a pandas dataframe from the data
df = pd.DataFrame(data = data.data, columns = data.feature_names)
df.head()
```

```python
#the above dataset doesnot contain the target values from the actual dataset
#now Attaching the target to the New data_frame
df['Target'] = data.target
#grttinh the overall view of New_df
df.head()
```

```python
#Exporting the Data_set so obtained in a JSON file
df.to_json("Breast_cancer_dataset.json")
```

## 4. Make a regression dataset (500) with 7 features while having 4 informative features and store them on disk in a csv file

```python
#We are naming our dataset as Sample_Df
#creating 7 features
Features = ['F1','F2','F3','F4','F5','F6','F7']

from sklearn import datasets

#creating a random dataset
Sample_x,Sample_y = datasets.make_regression(n_samples=500,n_features=7,n_informative=4,random_state=2)

#creating a pandas dataframe  
Sample_Df = pd.DataFrame(data = Sample_x,columns = Features)

#attaching the target values to the dataset
Sample_Df['Target'] = Sample_y
```

```python
#printing top 5 headers
Sample_Df.head()
```

```python
#saving the data into csv format
Sample_Df.to_csv("Regression_Dataset.csv")
```
