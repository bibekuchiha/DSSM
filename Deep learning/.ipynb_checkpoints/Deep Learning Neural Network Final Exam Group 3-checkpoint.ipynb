###### Dataset Description:
Honey is the natural product made by bees—one of our planet's most important species. Honey bees
visit millions of blossoms in their lifetimes, making pollination of plants possible and collecting nectar to
bring back to the hive to produce honey.
Honey bees use honey as their primary energy source and their instinct is to make more than their
colony needs. Beekeepers harvest the excess and bottle it for consumption, just like they’ve been doing
since the beginning of time. Harvesting honey is good for the bees and also part of what makes
successful beekeeping businesses.
This dataset provides insights into honey production, demand, and supply across different states of
America.
We would like to predict the average cost of honey (As our dependent variable) against other features
listed in the dataset.

```python
#For now, we care about the total average price of honey per year.
avg_per_year = df.groupby('year')['average_price'].mean().reset_index()
print(avg_per_year)
```

After creating X, we will need to reshape it to get it into the right format, using this command:

```python
X = avg_per_year['year']
X = X.values.reshape(-1,1)
print(X)
```

```python
df.head()
```

Encoding the state column because it is a categorical value


```python
from sklearn.preprocessing import OneHotEncoder

#Create an instance of One-hot-encoder
enc=OneHotEncoder()
  
#Passing encoded columns
enc_data=pd.DataFrame(enc.fit_transform(df[['state']]).toarray())
  
#Merge with main
New_df=df.join(enc_data)
  
print(New_df)
```

```python
New_df.drop('state',axis=1, inplace=True)
```

```python
New_df.head()
```

```python
X = New_df.drop('average_price',axis=1)
y = New_df.average_price
X
```

### Standardize the data


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()


df_std = scaler.fit_transform(X)
data = pd.DataFrame(df_std, columns=X.columns)
data.head()
```

```python
X = data
X
```

### Splitting into Train and Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X,y,test_size = 0.3, random_state = 0)
```

```python
print('Shape of X: ',X.shape)
```

```python
print('Shape of y: ',y.shape)
```

### Create and fit Linear Regression Model

```python
from sklearn.linear_model import LinearRegression


#Create a linear regression model from scikit-learn and call it regr
regr = LinearRegression()
#Fit the model to the data
regr.fit(X_train,y_train)
```

```python
print(' The coefficient is:' , regr.coef_)
print('The intercept is: ',regr.intercept_)
```

```python
y_predict = regr.predict(X_test)
```

```python
#Results  Lineer Regression
from sklearn import metrics
from sklearn.metrics import accuracy_score

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_predict))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_predict))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
```

### Create and fit Random Forest Regressor

```python
# Fitting Random Forest Regression to the dataset
# import the regressor
from sklearn.ensemble import RandomForestRegressor
  
 # create regressor object
regressor = RandomForestRegressor(n_estimators = 200, random_state = 0)
  
# fit the regressor with x and y data
regressor.fit(X_train, y_train)  
```

```python
y_predict = regressor.predict(X_test)
```

```python
print('The Accuracy is : ',regressor.score(X_test, y_test))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_predict))  
```

### Creating Deep sequential Neural Network

```python
from keras.models import Sequential
model = Sequential()
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

#first hidden layer
model = Sequential()
model.add(Dense(46, input_shape=(50,), kernel_initializer='normal', activation='relu'))
# Second hidden layer
model.add(Dense(20,activation='relu'))
# Output layer
model.add(Dense(1,activation='sigmoid'))
# Compiling the model
model.compile(optimizer='adam', loss='mae',metrics = ['mae'])
# Training our model for 50 epochs
model.fit(X_train,y_train,epochs=50)
```

```python
model.summary()
```

```python
# Train and Test accuracy
scores = model.evaluate(X_train,y_train)
print("Training Accuracy: " ,scores)
scores = model.evaluate(X_test,y_test)
print("Testing Accuracy:   ",scores)
```
